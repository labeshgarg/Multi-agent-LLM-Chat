```python
import autogen

BASE_URL="http://localhost:11434/v1"

config_list_mistral = [
    {
        'base_url': BASE_URL,
        'api_key': "fakekey",
        'model': "mistral",
    }
]

config_list_gemma7b = [
    {
        'base_url': BASE_URL,
        'api_key': "fakekey",
        'model': "gemma:7b",
    }
]

llm_config_mistral={
    "config_list": config_list_mistral,
}

llm_config_gemma7b={
    "config_list": config_list_gemma7b,
}
```

```python
user_proxy = autogen.UserProxyAgent(
    name="User_proxy",
    system_message="after first round writer is only allowed to message after editor's message",
    code_execution_config={"work_dir": "groupchat"},
    human_input_mode="TERMINATE",
    is_termination_msg=lambda x: x.get("content", "") and x.get("content", "").rstrip().endswith("TERMINATE"),
)
writer = autogen.AssistantAgent(
    name="Writer",
    system_message="""You are a blog post writer who is capable of writing a travel blog.
                      You generate one iteration of an article once at a time.
                      You wait for editor's change recommendation before generating another article.
                      You never provide review comments.
                      You are open to comments and willing to make changes in your article based on these comments by user_proxy.
                    """,
    llm_config=llm_config_mistral,
    max_consecutive_auto_reply=4,
    code_execution_config={"work_dir": "groupchat"},
    description="""This is a blog post writer who is capable of writing travel blogs.
                   The writer is open to any comments and recommendations from user_proxy for improving the article.
                   Ask writer to iterate article every time when there is a new change recommendation from user_proxy.
                """
)
editor = autogen.AssistantAgent(
    name="Editor",
    system_message="""You review blog posts and give change recommendations to writer to make the article more viral on social media, for example, adding hashtag.
                      You never write or revise blogs by yourself.
                      Reply TERMINATE if you are satisfy with the blog.
                       """,
    llm_config=llm_config_gemma7b,
    max_consecutive_auto_reply=3,
    code_execution_config={"work_dir": "groupchat"},
    description="""This is an editor who reviews the blogs of writer and provides change ideas to writer.
                    The editor should be called every time the writer provides a version of a blog post.
                """
)
groupchat = autogen.GroupChat(agents=[user_proxy, writer, editor], messages=[], max_round=6)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config_gemma7b and llm_config_mistral)
```

```python
user_proxy.initiate_chat(manager, message="""
Generate a 150-word of blog post about traveling in India.
""")
```

```python
for name in groupchat.messages:
    for key, value in name.items():
        if key=='name'or key=='content':
            print(f"{key}: {value}")
            
```
